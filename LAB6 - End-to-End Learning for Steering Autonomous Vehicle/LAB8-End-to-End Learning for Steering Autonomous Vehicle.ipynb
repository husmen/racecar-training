```python
# In order to train our network, we will also have an entry-less picture output
# we transfer them to a list
# When the camera starts to take the first shot, we do not get 30 records from the beginning.
images = list(df.FileName[30:])
labels = list(df.Angle[30:])
```

```python
# We make a histogram of how the distribution of the angles in our product is
#The fact that the distribution is not evenly causes the training to be improper
plt.hist(labels)
plt.show()
```

```python
# To correct some of the angle distribution in the dataset
# re-adding the records with the fewer number of points to the list
# we are trying to make it more smooth

nitem = len(images)
for i in range(nitem):
    if labels[i] > 0.05:
        for j in range(7):
            images.append(images[i])
            labels.append(labels[i])    
    if labels[i] < -0.07:
        for j in range(2):
            images.append(images[i])
            labels.append(labels[i]) 
```

```python
# According to the first histogram, we have reached a distribution that can be counted as balanced
# Not the right solution but a practical alternative
plt.hist(labels)
plt.show()
```

```python
# Settings related to our data set
# Data set batch size batch size
# How much training should be used for testing
# Education 80%, Test 20%
bsize = 8
dlen = len(labels)
splitpoint = int(0.8*dlen)
reindex = list(range(len(labels)))
# Eğtim verisini karıştıryoruz
random.seed(1234)
random.shuffle(reindex)
```

```python
# A function that applies a random brightness change on the image
# Augmentation function (taken from github)
def augment_brightness(image):
    image1 = cv2.cvtColor(image,cv2.COLOR_BGR2HSV) 
    image1 = np.array(image1, dtype = np.float64)
    random_bright = .5+np.random.uniform()
    image1[:,:,2] = image1[:,:,2]*random_bright
    image1[:,:,2][image1[:,:,2]>255]  = 255
    image1 = np.array(image1, dtype = np.uint8)
    image1 = cv2.cvtColor(image1,cv2.COLOR_HSV2RGB)
    return image1
```

```python
# read the name given picture
# apply a function that applies a brightness variation of 50% at random
# term function of the image matrix

def get_matrix(fname):
    img = cv2.imread(fname)
    if random.randint(0,1) == 1 :
        img = augment_brightness(img)        
    return img
```

```python
# It is not possible to get all the data in memory
# In addition, we can sometimes apply various modifications - Augmentation -
# With python generator, read data as needed and edit it for training or testing
# given to the system
# the underlying functions do this

# Generate data for training
def generate_data():
    i = 0
    while True:
        x = []
        y = []
        for j in range(i,i+bsize):  
            ix = reindex[j]
            img = get_matrix(images[ix])
            lbl = np.array([labels[ix]])
            flip = random.randint(0,1)
            if flip == 1:
                img = cv2.flip(img,1)
                lbl = lbl*-1.0
            x.append(img)
            y.append(lbl)
        x = np.array(x)
        y = np.array(y)       
        yield (x,y)    
        i +=bsize
        if i+bsize > splitpoint:
            i = 0
            
# Generate data for validation                  
def generate_data_val():
    i = splitpoint
    while True:
        x = []
        y = []
        for j in range(i,i+bsize): 
            ix = reindex[j]
            x.append(get_matrix(images[ix]))
            y.append(np.array([labels[ix]]))
        x = np.array(x)
        y = np.array(y)       
        yield (x,y)    
        i +=bsize
        if i+bsize > dlen:
            i = splitpoint
```

```python
# What you need for Keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Lambda
from keras.layers import Conv2D, MaxPooling2D, Cropping2D
from keras.callbacks import ModelCheckpoint
from keras.optimizers import SGD
```

```python
# Model based on NVIDIA's End to End Learning for Self-Driving Cars model
# We recognize a sequential keras model
model = Sequential()
# Cropping
# we write this layer to crop the image we want
# Cropping2D(cropping((top_crop, bottom_crop), (left_crop, right_crop)))
# in the following line 
# from top 144 to bottom 76 pixels
# truncating 672 pixels from 0 to the left
model.add(Cropping2D(cropping=((144,76),(0,672)), input_shape=(376,1344,3)))
# Normalization
# Values from 0 to 255 are scaled from -1 to 1
model.add(Lambda(lambda x: (2*x / 255.0) - 1.0))
# Convolution layer (5, 5) slides 24 by 2 sharps
model.add(Conv2D(24, (5, 5), activation="relu", strides=(2, 2)))
model.add(Conv2D(36, (5, 5), activation="relu", strides=(2, 2)))
model.add(Conv2D(48, (5, 5), activation="relu", strides=(2, 2)))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(Conv2D(64, (3, 3), activation="relu"))
#The network output is translated here
model.add(Flatten())
# Artificial Neural network part
model.add(Dense(100))
model.add(Dense(50))
model.add(Dense(10))
# Angle of exit of the network 
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
# The network we identified
model.summary()
```

```python
# We multiply the angle values ​​by 3 to get within the range of -0.3 to 0.3 and -1 to 1
labels = 3*np.array(labels)
```

```python
# During training, a function with the lowest value of the test error value
model_checkpoint = ModelCheckpoint('weights.h5', monitor='val_loss', save_best_only=True)
```

```python
# Training function 
hs = model.fit_generator(generate_data(),steps_per_epoch=int(splitpoint/ bsize),
                    validation_data=generate_data_val(), 
                    validation_steps=(dlen-splitpoint)/bsize, epochs=10,callbacks=[model_checkpoint])
```

```python
# We see graphically the course of the training process
# Train and validation loss chart
print(hs.history.keys())

plt.plot(hs.history['loss'])
plt.plot(hs.history['val_loss'])
plt.title('model mean squared error loss')
plt.ylabel('mean squared error loss')
plt.xlabel('epoch')
plt.legend(['training set', 'validation set'], loc='upper right')
plt.show()
```

```python
# We're recording the model we're training
# Network configuration in json
# Network parameter values ​​with an extension of h5
import json 
# Save model weights and json.
mname = 'model_new'
model.save_weights(mname+'.h5')
model_json  = model.to_json()
with open(mname+'.json', 'w') as outfile:
    json.dump(model_json, outfile)
```
